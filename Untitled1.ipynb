```python
import matplotlib.pyplot as plt
import numpy as np
x=np.arange(10)
plt.plot(x,x,"r--")
plt.plot(x,x**2,"bs")
plt.plot(x,x**3,"g^")
plt.xlabel("abs")
plt.ylabel("ordonnees")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.arange(10)
plt.plot(x,x,"r--",label="x")
plt.plot(x,x**2,"g",label="x^2")
plt.plot(x,x**3,"b",label="x^3")
plt.xlabel("abscisses")
plt.ylabel("ordonnees")
plt.title("Fonctions polynomiales")
plt.legend()
plt.show()
```

```python
plt.subplot(121) #la specification de la sous courbe doit être faite en spécifiant 3 chiffres
plt.plot(x,x,"r--",label="x")
plt.plot(x,x**2,"b",label="x^2")
plt.plot(x,x**3,"g",label="x^3")
plt.xlabel("abscisses")
plt.ylabel("ordonnees")
plt.title("Fonctions polynomiales 1-->3")
plt.legend()

plt.subplot(122)#LIGNE;COLONNE;INDEX 
plt.plot(x,x**4,"r--",label="x^4")
plt.plot(x,x**5,"b",label="x^5")
plt.plot(x,x**6,"g",label="x^6")
plt.xlabel("abscisses")
plt.ylabel("ordonnees")
plt.legend()
plt.title("Fonctions polynomiales 4-->6")
plt.show()
```

```python
rng = np.random.RandomState(0)
for marker in ['o', '.', ',', 'x', '+', 'v', '^', '<', '>', 's', 'd','*']:
   plt.plot(rng.rand(5), rng.rand(5), marker,
   label="marker='{0}'".format(marker))
   plt.legend(numpoints=1)
   plt.xlim(0, 1.8)
plt.show()
```

```python
rng = np.random.RandomState(0)
x = rng.randn(100)
y = rng.randn(100)
colors = rng.rand(100)
sizes = 1000 * rng.rand(100)

plt.scatter(x, y, c=colors, s=sizes, alpha=0.3,cmap='viridis')#pour comprendre les donnnees (densite)
plt.colorbar(); # show color scale
#plt.savefig('mes_courbes.png')
plt.show()
```
